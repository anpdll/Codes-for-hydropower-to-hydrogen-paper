```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv("46 degree celsius ui..csv",names = ['current','voltage']);
b = pd.read_csv("72 degree celsius ui..csv",names = ['current','voltage']);
c = pd.read_csv("59 degree celsius ui..csv",names = ['current','voltage']);
```

```python
a['voltage']= a['voltage']/75;
a['current'] =  a['current']/2082;

b['voltage']= b['voltage']/75;
b['current'] =  b['current']/2082;

c['voltage']= c['voltage']/75;
c['current'] =  c['current']/2082;
```

```python
a = a[a['voltage']>=1.6];
b = b[b['voltage']>=1.6];
c = c[c['voltage']>=1.6];
```

```python
a.to_csv('46_degree_corrected.csv',index=False);
b.to_csv('72_degree_corrected.csv',index=False);
c.to_csv('59_degree_corrected.csv',index=False);
```

```python
plt.plot(a['current'],a['voltage'])
plt.plot(b['current'],b['voltage'])
plt.plot(c['current'],c['voltage'])
```
